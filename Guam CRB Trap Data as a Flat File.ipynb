### Create a flat file by combining location data from the *trap* table and trap catch data from the *trap_visit* table

```python
connection_string = "mysql+pymysql://{}:{}@mysql.guaminsects.net/oryctes".format(username, password)
engine = create_engine(connection_string)
cnx = engine.raw_connection()
sql = '''
SELECT trap.trap_id, trap.date_deployed, trap.date_removed, trap.latitude, trap.longitude,
    trap_visit.visit_date, 
    (trap_visit.male_count+trap_visit.female_count+trap_visit.unsexed_count) AS beetles_trapped
FROM trap, trap_visit
WHERE trap.trap_id = trap_visit.trap_id
LIMIT 100000;
'''
data = pd.read_sql(sql, cnx)
data
```

### Save the flat file as *trap_data.csv*.

```python
data.to_csv('trap_data.csv', index=False)
```
